```python
def calculate_rsi(data, window=50):
    """
    Calculates the Relative Strength Index (RSI) for a given dataset.
    
    :param data: pandas Series of closing prices.
    :param window: The period for RSI calculation (default is 14).
    :return: pandas Series with RSI values.
    """
    # 1. Calculate price changes
    delta = data.diff(1)
    
    # 2. Separate gains and losses
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    
    # 3. Calculate the smoothed average of gains and losses
    # The first value is a simple average, subsequent values are exponentially weighted.
    # com=window-1 is equivalent to alpha=1/window, which is the standard for RSI.
    avg_gain = gain.ewm(com=window - 1, adjust=False).mean()
    avg_loss = loss.ewm(com=window - 1, adjust=False).mean()
    
    # 4. Calculate Relative Strength (RS)
    rs = avg_gain / avg_loss
    
    # 5. Calculate the RSI
    rsi = 100 - (100 / (1 + rs))
    
    return rsi
```

```python
df['RSI'] = df.groupby('Symbol')['Close'].transform(lambda x: calculate_rsi(x, window=50))
```

```python
def calculate_macd(data, short_window=12, long_window=26, signal_window=9):
    """
    Calculate MACD (Moving Average Convergence Divergence).
    
    Parameters:
        data          : pandas Series of prices (e.g. closing prices)
        short_window  : EMA short period (default=12)
        long_window   : EMA long period (default=26)
        signal_window : EMA of MACD line (default=9)
        
    Returns:
        macd_line, signal_line, histogram : pandas Series
    """
    # Short-term EMA
    ema_short = data.ewm(span=short_window, adjust=False).mean()
    
    # Long-term EMA
    ema_long = data.ewm(span=long_window, adjust=False).mean()
    
    # MACD Line
    macd_line = ema_short - ema_long
    
    # Signal Line
    signal_line = macd_line.ewm(span=signal_window, adjust=False).mean()
    
    # Histogram
    histogram = macd_line - signal_line
    
    return macd_line, signal_line, histogram



data =df.Close
macd_line, signal_line, histogram = calculate_macd(data)

print("MACD Line:\n", macd_line)
print("\nSignal Line:\n", signal_line)
print("\nHistogram:\n",histogram)
```
